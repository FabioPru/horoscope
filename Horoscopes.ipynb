```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
import csv

signs = ["aries", "taurus", "gemini", "cancer", "leo", "virgo", "libra", "scorpio", "sagittarius", "capricorn", "aquarius", "pisces"]
months = ["january", "february", "march", "april", "may", "june", "july", "august", "september", "october", "november", "december"]
days = []
for day in range(1, 32):
    days.append(str(day))
years = ["2018"]
date_a = []
sign_a = []
text_a = []
for sign in signs:
    for month in months:
        for day in days:
            for year in years:
                try: 
                    print(9)
                    b = "https://mystarshoroscopes.com/" + sign + "-daily-horoscope-for-" + month + "-" + day + "-" + year + "/"
                    requests.get(b, verify=True)
                    html = requests.get(b).text
                    soup = BeautifulSoup(html, 'html5lib')
                    a = soup.find(class_ = "entry-content")
                    if a:
                        a = a.text
                        if a != "The page you are looking for no longer exists. Perhaps you can return back to the homepage and see if you can find what you are looking for. Or, you can try finding it by using the search form below.":
                            text_a.append(a)
                            sign_a.append(sign)
                            date_a.append(day + " " + month + " " + year)
                except:
                    print(0)
                        
df = pd.DataFrame({"Date": date_a , "Sign": sign_a  , "Text": text_a }) 
export_csv = df.to_csv (r'C:\Users\Giuliano\Documents\horoscope_2018.csv', index = None, header=True)
```
